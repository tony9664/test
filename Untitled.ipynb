```python
import pandas as pd
data=pd.read_csv('Fremont.csv',index_col='Date',parse_dates=True) 
data.head()
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')
data.columns=['West','East']
data.resample('W').sum().plot()
```

```python
ax=data.resample('D').sum().rolling(365).sum().plot()
ax.set_ylim(0,None)
```

```python
data['Total'] = data['West']+data['East']
```

```python
data.groupby(data.index.time).mean().plot()
```

```python
pivoted = data.pivot_table('West',index=data.index.time,columns=data.index.date)
pivoted2 = data.pivot_table('East',index=data.index.time,columns=data.index.date)
pivoted.iloc[:5,:5]
```

```python
pivoted.plot(legend=False,alpha=0.01,c='r')
pivoted2.plot(legend=False,alpha=0.01,c='b')
```
